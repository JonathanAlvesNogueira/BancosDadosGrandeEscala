```python
import pandas as pd
import numpy as np
import random 
tamanho_base = 400000
```

# Gera a Base do Santander - Visão Produtos

```python
# define quantos registros vai ter a base

# Gera o tipo de pessoa aleatoriamente
tipo_pessoa = np.random.choice(['PF', 'PJ'], size=tamanho_base)

#  gera o cpf_cnpj aleatoriamente
def gera_cpf_sem_dac():
    cpf = ''.join([str(random.randint(0, 9)) for i in range(9)])
    return cpf

cpf_cnpj = [ gera_cpf_sem_dac() for i in range(tamanho_base)]

renda_faturamento = np.random.choice([np.random.randint(1000, 10000), np.random.randint(10000,50000), np.random.randint(50000, 100000)], size=tamanho_base, p=[0.7, 0.2, 0.1])

outro_banco_cliente = np.random.choice(['Nubank', 'C6', 'Bradesco', 'Santander', 'Itau', 'Inter'], size=tamanho_base)

bad30 = np.random.randint(0, 2, size=tamanho_base)

serasa_score = np.random.randint(0, 999, size=tamanho_base)

tipo_bem = np.random.choice(['Veículos', 'Motos', 'Imóveis', 'Pesados'], size=tamanho_base)

valor_do_bem = [np.random.randint(5000, 10000) if bem == 'Motos' else np.random.randint(50000, 200000) if bem == 'Veículos' else np.random.randint(100000, 500000) if bem == 'Imóveis' else 
                np.random.randint(50000, 300000) for bem in tipo_bem]



fl_cliente_outro_produto = np.random.randint(0, 2, size=tamanho_base)


dias_atraso_todos_produtos = np.random.choice( [0, np.random.randint(0,540)], size=tamanho_base, p=[0.8,0.2])

aprovados = np.random.choice(['Aprovado pelo Motor', 'Aprovado por Analista'], size=tamanho_base, p=[0.8, 0.2])

fl_recebe_salario_banco = np.random.choice([0, 1], size=tamanho_base, p=[0.6, 0.4])
```

```python
# Junta todas as colunas de 10.000.000 de clientes que possuem Produtos Financeiros no Santander (Consorcio, Financiamento)
data = {
    'CPF_CNPJ': cpf_cnpj,
    'Renda_faturamento': renda_faturamento,
    'Outro_banco_cliente': outro_banco_cliente,
    'Bad30': bad30,
    'Serasa_score': serasa_score,
    'Tipo_bem': tipo_bem,
    'Valor_do_bem': valor_do_bem,
    'FL_cliente_outro_produto': fl_cliente_outro_produto
    ,'TIPO_PESSOA': tipo_pessoa
    ,'DIAS_ATRASO': dias_atraso_todos_produtos
    ,'Desc_cliente':aprovados
    ,'fl_recebe_salario_banco': fl_recebe_salario_banco
}

df_info = pd.DataFrame(data)
display(df_info)
```
